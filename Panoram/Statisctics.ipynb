```python
proteins = parse_proteins_in_file("Data/de_proteins.txt")
```

```python
data = pd.read_csv("Data/smears_imputed.csv", index_col=0, header=[0, 1, 2,])
data = np.log10(data.T.reset_index().iloc[:, 2:].set_index("Sample") + 1)
# data = data.T.reset_index().iloc[:, 2:].set_index("Sample")

meta = pd.read_csv("Data/sample_params.csv", index_col=0)

from sklearn.preprocessing import LabelEncoder

enc = LabelEncoder()

encoded_y = enc.fit_transform(meta["condition"].loc[data.index])
encoded_y

data["condition"] = encoded_y

params = [ i for i in data.columns if i != 'condition']
```

```python
logit_model = statsmodels.discrete.discrete_model.Logit.from_formula(formula=f"condition ~ {' + '.join(proteins)}", data=data)
result=logit_model.fit()
print(result.summary())
```

```python
data = pd.read_csv("Data/smears_imputed.csv", index_col=0, header=[0, 1, 2,])
data = np.log10(data.T.reset_index().iloc[:, 2:].set_index("Sample") + 1)
# data = data.T.reset_index().iloc[:, 2:].set_index("Sample")

meta = pd.read_csv("Data/sample_params.csv", index_col=0)

from sklearn.preprocessing import LabelEncoder

enc = LabelEncoder()

encoded_y = enc.fit_transform(meta["acuteness"].loc[data.index])
encoded_y

data["acuteness"] = encoded_y

params = [ i for i in data.columns if i != 'acuteness']
```

```python
logit_model = smf.logit(f"acuteness ~ {' + '.join(params[1:])}", data=data)
result=logit_model.fit()
print(result.summary())
```
