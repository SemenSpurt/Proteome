```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
from matplotlib.patches import Patch

def plot_distributions(references, samples):
    fig, ax = plt.subplots(figsize=(6, 4))

    sns.kdeplot(references, ax=ax, linewidth=2, color="black", alpha=0.75, label="Контрольный образец")
    sns.kdeplot(samples, ax=ax, linewidth=1, color="blue", fill=True, alpha=0.5, label="Исследуемые образцы")

    plt.legend(
        
        [
            Patch(color='black',   alpha=1), 
            Patch(color='blue',    alpha=0.5), 
        ], 
        
        [
            "Внутренний стандарт", 
            "Исследуемые образцы", 
        ], 
        loc="upper left",

    )

    plt.title(protein, fontsize=12)
```

```python
from datetime import datetime

tab = pd.read_csv("/home/selezov_su/Proteomics/Norma/Data/capillary_proteome_factorized_dataset.csv", parse_dates=["Collection_Date"])
values = tab.pivot_table(index="Protein_ID", columns=["Preparation_Batch", "Sample"], values="Area").apply(np.log10)

meta = tab[["Preparation_Batch", "Sample", "Patient", "Collection_Date", "Collection_Time", "Sample_Type"]].drop_duplicates().reset_index(drop=True)
meta.columns = ["batch", "sample", "patient", "date", "time", "type"]
meta.to_csv("/home/selezov_su/notebooks/MRM/data/meta.csv", index=False)
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import SimpleImputer, IterativeImputer, KNNImputer

imputer = IterativeImputer(max_iter=2, min_value=values.min().min())
filled = pd.DataFrame(imputer.fit_transform(values), index=values.index, columns=values.columns)

fig = plt.figure(figsize=(40, 5))
plt.boxplot(filled)
plt.show()

from combat.pycombat import pycombat

batch = [c[0] for c in filled.columns]
corrected = pycombat(filled, batch)

fig = plt.figure(figsize=(40, 5))
plt.boxplot(corrected)
plt.show()
```

```python
references = corrected[[c for c in corrected.columns if c[1].startswith("Ref")]]
samples = corrected[[c for c in corrected.columns if c[1].startswith("OK")]]

samples.to_csv("/home/selezov_su/notebooks/MRM/data/samples.csv")
references.to_csv("/home/selezov_su/notebooks/MRM/data/references.csv")
```
